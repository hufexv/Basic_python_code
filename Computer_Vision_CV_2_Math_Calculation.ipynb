```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
#paste
img = cv.imread('person.png')
pixes1 = img[0:95, 20:240]
img[0:95, 280:500] = pixes1
cv2_imshow(img)
```

```python
#split of kernel
b, g, r = cv.split(img)
cv2_imshow(np.stack([b,g,np.zeros_like(b)], axis=2))
#concat stack
cv2_imshow(cv.merge((b,g,r)))
```

```python
#add_border
img = cv.imread('OpenCV.png')
#replicate
replicate = cv.copyMakeBorder(img, 10, 10, 10, 10, cv.BORDER_REPLICATE)
#reflect
reflect = cv.copyMakeBorder(img, 10, 10, 10, 10, cv.BORDER_REFLECT)
#reflect101
reflect101 = cv.copyMakeBorder(img, 10, 10, 10, 10, cv.BORDER_REFLECT_101)
#wrap
wrap = cv.copyMakeBorder(img, 10, 10, 10, 10, cv.BORDER_WRAP)
#constant
constant = cv.copyMakeBorder(img, 10, 10, 10, 10, cv.BORDER_CONSTANT, value=[255,255,0])

#
plt.subplot(231)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title('original')
plt.subplot(232)
plt.imshow(cv.cvtColor(replicate, cv.COLOR_BGR2RGB))
plt.title('replicate')
plt.subplot(233)
plt.imshow(cv.cvtColor(reflect, cv.COLOR_BGR2RGB))
plt.title('reflect')
plt.subplot(234)
plt.imshow(cv.cvtColor(reflect101, cv.COLOR_BGR2RGB))
plt.title('reflect101')
plt.subplot(235)
plt.imshow(cv.cvtColor(wrap, cv.COLOR_BGR2RGB))
plt.title('wrap')
plt.subplot(236)
plt.imshow(cv.cvtColor(constant, cv.COLOR_BGR2RGB))
plt.title('constant')
plt.show()
```

```python
#add_border(array)
arr = np.arange(9).reshape((3,3))

replicate = cv.copyMakeBorder(arr, 3, 3, 3, 3, cv.BORDER_REPLICATE)
reflect = cv.copyMakeBorder(arr, 3, 3, 3, 3, cv.BORDER_REFLECT)
reflect101 = cv.copyMakeBorder(arr, 3, 3, 3, 3, cv.BORDER_REFLECT_101)
wrap = cv.copyMakeBorder(arr, 3, 3, 3, 3, cv.BORDER_WRAP)
constant = cv.copyMakeBorder(arr, 3, 3, 3, 3, cv.BORDER_CONSTANT, value=0)

print(replicate)
print(reflect)
print(reflect101)
print(wrap)
print(constant)
```

```python
#combine of img
img1 = cv.imread('person.png')
img2 = cv.imread('OpenCV.png')

#resize
img1 = cv.resize(img1, (300, 300))
img2 = cv.resize(img2, (300, 300))

img = cv.addWeighted(src1=img1, alpha=0.3, src2=img2, beta=1.0, gamma=0)
cv2_imshow(img)
```

```python
#combine of img(array)
a = img1 * 0.3 +img2 * 1.0
print(np.max(a))
a = np.clip(a, a_min=0, a_max=255).astype(np.uint8)
cv2_imshow(a)
```

```python
#img_preprocess
img = cv.imread('OpenCV.png')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_blur = cv.GaussianBlur(img_gray, (3,3), 0)
img_ret, img_mask = cv.threshold(img_gray, 10, 255, cv.THRESH_BINARY)
mask_inv = cv.bitwise_not(img_mask)

cv2_imshow(img_gray)
cv2_imshow(img_blur)
cv2_imshow(img_mask)
cv2_imshow(mask_inv)
```
