```python
import os
import torch
from PIL import Image
from torchvision import models, transforms, utils


def t0():
  os.makedirs("./output", exist_ok="True")
  net = models.vgg16(weights=models.VGG16_Weights.DEFAULT)
  net = net.eval()
  print(net)
  #load_image
  path = "./person.jpg"
  img: Image.Image = Image.open(path)
  print(img.size)
  img = img.resize((800,500))
  img = img.convert("RGB")
  print(img.size)

  compose = transforms.Compose(
      transforms=[
          transforms.ToTensor()
      ]
  )
  img_tensor = compose(img)
  print(img_tensor.shape)

  #train
  in_feature = img_tensor[None]
  out_score = net(in_feature)
  print(out_score.shape)
  #get_top5_possible_class
  out_prob = torch.softmax(out_score, dim=-1)
  topk_prob_values, topk_class_indices = torch.topk(out_prob, k=5, dim=-1)
  print(f"predict_class_pro{topk_prob_values}")
  print(f"predict_class_id{topk_class_indices}")
  #makedir
  fp_output_dir = os.path.join("./output", os.path.splitext(os.path.basename(path))[0])
  os.makedirs(fp_output_dir, exist_ok=True)
  print(f"output_file{fp_output_dir}")
  #save_each_layer_image
  layer_out_feature = net.features[0](in_feature)
  print(layer_out_feature.shape)
  utils.save_image(
      tensor=torch.transpose(layer_out_feature, 0, 1),
      fp=f"{fp_output_dir}/layer_1.png"
  )

  weight = net.features[0].weight
  bias = net.features[0].bias
  print(weight.shape)
  print(bias.shape)
  print(weight[0])
  print(bias[0])

  in_feature = layer_out_feature
  layer_out_feature = net.features[1](in_feature)
  print(layer_out_feature.shape)
  utils.save_image(
      tensor=torch.transpose(layer_out_feature, 0, 1),
      fp=f"{fp_output_dir}/layer_2.png"
  )
  in_feature = layer_out_feature
  layer_out_feature = net.features[2](in_feature)
  print(layer_out_feature.shape)
  utils.save_image(
      tensor=torch.transpose(layer_out_feature, 0, 1),
      fp=f"{fp_output_dir}/layer_3.png"
  )
  in_feature = layer_out_feature
  layer_out_feature = net.features[3](in_feature)
  print(layer_out_feature.shape)
  utils.save_image(
      tensor=torch.transpose(layer_out_feature, 0, 1),
      fp=f"{fp_output_dir}/layer_4.png"
  )
  in_feature = layer_out_feature
  layer_out_feature = net.features[4](in_feature)
  print(layer_out_feature.shape)
  utils.save_image(
      tensor=torch.transpose(layer_out_feature, 0, 1),
      fp=f"{fp_output_dir}/layer_5.png"
  )
  #整个特征提取图
  in_feature = img_tensor[None]
  layer_out_feature = net.features(in_feature)
  print(layer_out_feature.shape)
  utils.save_image(
      tensor=torch.transpose(layer_out_feature, 0, 1),
      fp=f"{fp_output_dir}/layer_all.png",
      pad_value=0.5
  )

  print("success")
```

```python
if __name__ == '__main__':
  t0()
```
